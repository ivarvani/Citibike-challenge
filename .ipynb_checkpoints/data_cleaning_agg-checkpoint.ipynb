```python
# Dependencies
import pandas as pd
```

```python
# List of file names to be combined 
files20 = ['data/202001_citibike_tripdata.csv', 'data/202002_citibike_tripdata.csv', 'data/202003_citibike_tripdata.csv',
           'data/202004_citibike_tripdata.csv', 'data/202005_citibike_tripdata.csv', 'data/202006_citibike_tripdata.csv',
           'data/202007_citibike_tripdata.csv','data/202008_citibike_tripdata.csv', 'data/202009_citibike_tripdata.csv',
           '202010_citibike_tripdata.csv','202011_citibike_tripdata.csv','202012_citibike_tripdata.csv',]
# Create an empty list to store each dataframe
df20_list = []

# Loop through each file in the list
for file in files20:
    # Read the file into a dataframe
    df20 = pd.read_csv(file)

    # Append the dataframe to the list
    df20_list.append(df20)

# Concatenate all dataframes in the list into one dataframe
df20 = pd.concat(df20_list)

# Show dataframe
df20
```

```python
# Dictionary mapping values in the 'gender' column to new values
mapping = {0: 'Unknown', 1: 'Male', 2: 'Female'}

# Use the 'map' method to change the values in the 'gender' column
df20['gender'] = df20['gender'].map(mapping)

# Show DataFrame
df20
```

```python
df20['Trip Year'] = df20.starttime.str.split("-", expand=True)[0]
df20.head()
```

```python
#break out month
df20['Trip Month'] = df20.starttime.str.split("-", expand=True)[1]
df20.head()

df20["Trip Month"].replace({"01": "January","02": "February","03": "March","04": "April","05": "May","06": "June","07": "July",
                            "08": "August", "09": "September", "10": "October","11": "November","12": "December"}, inplace=True) 
df20
```

```python
# #calculate age of rider

# df20["Rider Age"] = df20['Trip Year']-df20['birth year']

# df20.head()

df20['Trip Year']=df20['Trip Year'].astype("int64")
```

```python
df20.info()
```

```python
#calculate age of rider

df20["Rider Age"] = df20['Trip Year']-df20['birth year']

df20.tail()
```
